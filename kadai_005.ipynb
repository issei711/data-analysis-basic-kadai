```python
import numpy as np
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 3, tight_layout=True)

# [0,0]
y00 = [1, 2, -5, 2]
axes[0,0].plot(y00)

# [0,1]
x01 = np.linspace(0, 10, 100)
y01 = 2 + 2 * np.sin(2 * x01)
plot01 = axes[0,1]
plot01.plot(x01, y01, linewidth=2.0)
plot01.set(xlim=(0, 10), xticks=np.arange(0, 10),
           ylim=(-1, 5), yticks=np.arange(-1, 6))

# [0,2]
x02 = np.random .normal(15, 5, 5000)
plot02 = axes[0,2]
plot02.hist(x02, bins=30)
plot02.set_xlim(-5, 30)

# [1,0]
np.random.seed(3)
x10 = 4 + np.random.normal(0, 2, 60)
y10 = 4 + np.random.normal(0, 2 ,len(x10))
axes[1, 0].scatter(x10, y10)

plt.show()
```
